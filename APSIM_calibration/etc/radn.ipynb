```python
import pandas as pd
import numpy as np

# data = pd.read_csv('C:\code\Apsim_2023\penman-monteith\Gangneung_weather.csv', parse_dates=['날짜'])
# data = pd.read_csv('C:\code\Apsim_2023\penman-monteith\P-monteith.csv',parse_dates=['날짜'])
data = pd.read_csv("../output/weather_csv/Gangneung_weather.csv", parse_dates=['날짜'])
date = data['날짜']
doy = date.dt.dayofyear.values.tolist()
data.head(40)
```

```python
date = data['날짜'].values.tolist()
month = data['월'].values.tolist()
day = data['일'].values.tolist()


Tavg = data['평균기온'].values.tolist()
Hmin = data['평균습도'].values.tolist()
Windmin = data['평균풍속'].values.tolist()
daytime = data['일조시간'].values.tolist()

data['북위'] = 37.54
data['해발고도'] = 28

latitude = 37.54
altitude = 28

windheight = data['풍속계높이'].values.tolist()

u_2 = []
P = []
delta = []
gamma = []

u_2_cal = []
Q = []
R =[]
S = []

e_s =[]
e_a = []

e = []

dr = []
small_delta =  []
theta = []
w_s = []

Ra = []

N = []
Rs  = []
Rso = []
RsRso = []
R_ns = []
R_nl = []

Rn = []

Tavg
```

```python

for i in range(len(date)):
    u_2.append(Windmin[i] * 4.87/(np.log(67.8 * windheight[i] - 5.42)))
    P.append(101.3 * ((293 - 0.0065*altitude)/293)**5.26)
    delta.append(4098*(0.6108*np.exp((17.27*Tavg[i])/(Tavg[i]+237.3)))/(Tavg[i]+237.3)**2)
    gamma.append(0.665*10**-3*P[i])

    u_2_cal.append(1 + 0.34 * u_2[i])
    Q.append(delta[i] / (delta[i] + gamma[i] * u_2_cal[i]))
    R.append(gamma[i] / (delta[i] + gamma[i] * u_2_cal[i]))
    S.append(900 / (Tavg[i] + 273) * u_2[i])

    e_s.append(0.6108 * np.exp((17.27*Tavg[i])/(Tavg[i]+237.3)))
    e_a.append(Hmin[i]/100 * e_s[i])

    e.append(e_s[i] - e_a[i])

    dr.append(1+0.033*np.cos(2*3.141592/365*doy[i]))
    small_delta.append(0.409*np.sin(2*3.141592/365*doy[i]-1.39))
    theta.append(latitude * 3.141592/180)
    w_s.append(np.arccos(-np.tan(theta[i])*np.tan(small_delta[i])))

    Ra.append(24*60/3.141592*0.082*dr[i]*(w_s[i]*np.sin(small_delta[i])*np.sin(theta[i])+np.cos(theta[i])*np.cos(small_delta[i])*np.sin(w_s[i])))

    N.append(24/3.141592*w_s[i])
    Rs.append((0.25+0.5*daytime[i]/N[i])*Ra[i])
    Rso.append((0.75+2*10**-5*altitude)*Ra[i])
    RsRso.append(Rs[i] / Rso[i])
    R_ns.append(0.77*Rs[i])
    R_nl.append(4.903*10**-9*(Tavg[i]+273.16)**4*(0.34-0.14*e_a[i]**0.5)*(1.35*RsRso[i]-0.35))

    Rn.append(R_ns[i] - R_nl[i])
Ra
```
